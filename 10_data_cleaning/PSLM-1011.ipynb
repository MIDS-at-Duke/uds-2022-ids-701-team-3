```python
import pandas as pd

C = pd.read_stata(
    "/Users/johnowusuduah/Downloads/data  in stata 2010-2011/1011_sec_c_dist.dta"
)
print(C.shape)
C.sample(5)
```

```python
# Reduce to the questions of interest (4 questions):

C["ever_admitted"] = C["scq03"]
C["max_level_acheived"] = C["scq04"]
C["currently_enrrolled"] = C["scq05"]
C["why_not"] = C["scq91"]

C_relevant = C[
    [
        "hhcode",
        "idc",
        "ever_admitted",
        "max_level_acheived",
        "currently_enrrolled",
        "why_not",
        "region",
        "province",
    ]
].copy()
```

```python
C_relevant
```

```python
B = pd.read_stata(
    "/Users/johnowusuduah/Downloads/data  in stata 2010-2011/1011_sec_b_dist.dta"
)
B.sample(5)

```

```python
# Change the names of the questions for readability
B['sex'] = B['sbq03']
#B['age'] = B['sbq04']
B['marital_status'] = B['sbq06']

B_relevant = B[['hhcode', 'idc', 'age', 'sex', 'marital_status']].copy()
```

```python
# Made the Household Code column a str type
B["hhcode"] = B["hhcode"].astype("str")
C["hhcode"] = C["hhcode"].astype("str")
```

```python
# MERGE
years10_11 = pd.merge(B_relevant, C_relevant, on=["hhcode", "idc"], how='inner', indicator=True)

# Evaluate merge 
years10_11._merge.value_counts()
```

```python
# Select people in school enrollment age (I would say 4-15 years old)
years10_11 = years10_11.loc[(years10_11["age"] >= 4)&(years10_11["age"] <= 15)]
```

```python
years10_11.head()
```

```python
years10_11["marital_status"].value_counts()
```

```python
# Lets see which hh codes we have:
years10_11["hhcode"] = years10_11["hhcode"].astype("str")
years10_11['subprovince code']= years10_11['hhcode'].apply(lambda x: x[0:4])
# remove the decimal characters from the string values
years10_11['hhcode']= years10_11['hhcode'].apply(lambda x: x[0:-2])

years10_11["subprovince code"] = years10_11["subprovince code"].astype("int")
years10_11.loc[(years10_11["region"] == 1), 'region'] = 'urban'
years10_11.loc[(years10_11["region"] == 2), 'region'] = 'rural'
# reformat marital status values
years10_11["marital_status"] = years10_11["marital_status"].astype("str")
years10_11.loc[(years10_11["marital_status"] == "unmarried"), 'marital_status'] = 1
years10_11.loc[(years10_11["marital_status"] == "married"), 'marital_status'] = 2
years10_11.loc[(years10_11["marital_status"] == "widow"), 'marital_status'] = 3
years10_11.loc[(years10_11["marital_status"] == "divorced"), 'marital_status'] = 4
years10_11.loc[(years10_11["marital_status"] == "nikkah"), 'marital_status'] = 5
years10_11.sample(5)
```

```python
years10_11["marital_status"].value_counts()
```

```python
# Now we import the stratum name - code data
strat_name = pd.read_csv(
    "/Users/johnowusuduah/Downloads/data  in stata 2010-2011/stratum1011.csv"
).iloc[:,0:3]

strat_name.sample(5)
```

```python
# MERGE
years10_11_ = pd.merge(years10_11, strat_name, on=["subprovince code"], how='inner')

# Evaluate merge 
years10_11_._merge.value_counts()
```

```python
years10_11_.sample(5)
```

```python
# FINAL DATA PRE-PROCESSING
# subset columns of interest
years10_11_f = years10_11_[["hhcode","age","idc","sex","marital_status","ever_admitted","currently_enrrolled","region","subprovince code","province_y","subprovince"]].copy()
# rename index and province variables
years10_11_f = years10_11_f.rename(columns={"province_y":"province"})
# add year to data
years10_11_f["year"] = 2010
# change NWFP to KPK
years10_11_f.loc[years10_11_f["province"] == "NWFP", "province"] = "KPK"
# change Khyber/P.K to KPK
years10_11_f.loc[years10_11_f["province"] == "Khyber/P.K", "province"] = "KPK"
# reformat columns/variables for consistency across all data sets
years10_11_f["sex"] = years10_11_f["sex"].astype("category")
years10_11_f["marital_status"] = years10_11_f["marital_status"].astype("int8")
years10_11_f["ever_admitted"] = years10_11_f["ever_admitted"].astype("str")
years10_11_f["currently_enrrolled"] = years10_11_f["currently_enrrolled"].astype("str")
years10_11_f["region"] = years10_11_f["region"].astype("str")
# make sure Islamabad is Punjab
years10_11_f.loc[(years10_11_f["province"] == "Islamabad"), 'province'] = "Punjab "
# preview sample of data set
years10_11_f.sample(5)
```

```python
import pandas as pd

C = pd.read_stata(
    "/Users/johnowusuduah/Downloads/data in stata 2008-09/sec_c.dta"
)
print(C.shape)
C.sample(5)
```

```python
# Reduce to the questions of interest (4 questions):

C["ever_admitted"] = C["scq03"]
C["max_level_acheived"] = C["scq04"]
C["currently_enrrolled"] = C["scq05"]
C["why_not"] = C["scq09a"]

C_relevant = C[
    [
        "hhcode",
        "idc",
        "ever_admitted",
        "max_level_acheived",
        "currently_enrrolled",
        "why_not",
        "region",
        "province",
    ]
].copy()
```

```python
C_relevant
```

```python
B = pd.read_stata(
    "/Users/johnowusuduah/Downloads/data in stata 2008-09/sec_b.dta"
)
B.sample(5)
```

```python
# Change the names of the questions for readability
B['sex'] = B['sbq01']
#B['age'] = B['sbq04']
B['marital_status'] = B['sbq05']

B_relevant = B[['hhcode', 'idc', 'age', 'sex', 'marital_status']].copy()
```

```python
# Made the Household Code column a str type
B["hhcode"] = B["hhcode"].astype("str")
C["hhcode"] = C["hhcode"].astype("str")
```

```python
# MERGE
years08_09 = pd.merge(B_relevant, C_relevant, on=["hhcode", "idc"], how='inner', indicator=True)

# Evaluate merge 
years08_09._merge.value_counts()
```

```python
# Select people in school enrollment age (I would say 4-15 years old)
years08_09 = years08_09.loc[(years08_09["age"] >= 4)&(years08_09["age"] <= 15)]
```

```python
years08_09.head()
```

```python
years08_09.province.value_counts()
```

```python
# Lets see which hh codes we have:
years08_09["hhcode"] = years08_09["hhcode"].astype("str")
years08_09['subprovince code']= years08_09['hhcode'].apply(lambda x: x[0:4])
years08_09["subprovince code"] = years08_09["subprovince code"].astype("int")
years08_09.head()
```

```python
# Now we import the stratum name - code data
strat_name = pd.read_csv(
    "/Users/johnowusuduah/Downloads/data in stata 2008-09/stratum0809.csv"
)
strat_name.sample(5)
```

```python
# MERGE
years08_09_ = pd.merge(years08_09, strat_name, on=["subprovince code"], how='inner')

# Evaluate merge 
years08_09_._merge.value_counts()
```
